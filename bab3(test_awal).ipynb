## 1. Membuat Database

Screenshot 2025-11-22 102027.png

Perintah `CREATE DATABASE` digunakan untuk membuat sebuah database baru di MySQL.  
Perintah ini menyiapkan ruang penyimpanan yang nantinya digunakan untuk tabel, data, dan struktur database lainnya.

**Fungsi utama:**
- Menginisialisasi database baru sebelum membangun tabel atau memasukkan data.

---

## 2. Melihat Seluruh Database pada MySQL Server
Screenshot 2025-11-22 102049.png

Perintah `SHOW DATABASES` digunakan untuk menampilkan seluruh database yang tersimpan pada server MySQL.  
Perintah ini digunakan untuk memastikan database yang telah dibuat muncul dan untuk melihat database bawaan MySQL.

**Fungsi utama:**
- Menampilkan daftar lengkap database yang tersedia.

---

## 3. Mengakses / Menggunakan Database

Screenshot 2025-11-22 102111.png

Sebelum membuat tabel atau menjalankan query pada suatu database, kita harus memilih database tersebut dengan perintah `USE`.

**Fungsi utama:**
- Mengatur konteks kerja agar semua perintah SQL berikutnya diterapkan pada database yang dipilih.

---

## 4. Menghapus Database

Screenshot 2025-11-22 102122.png

Perintah `DROP DATABASE` digunakan untuk menghapus sebuah database beserta seluruh tabel dan datanya secara permanen.  
Perintah ini harus digunakan dengan hati-hati karena tidak bisa dibatalkan.

**Fungsi utama:**
- Menghapus database yang sudah tidak dibutuhkan agar server tetap rapi dan hemat penyimpanan.
