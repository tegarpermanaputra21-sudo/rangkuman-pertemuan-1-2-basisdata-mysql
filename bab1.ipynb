## ğŸ”„ Aturan Konversi ERD â†’ Relasi

### **1. Entitas Kuat â†’ Tabel**
- Setiap entitas kuat menjadi tabel.
- Atribut â†’ kolom.
- PK tetap digunakan.

---

### **2. Atribut Komposit**
Atribut yang terdiri dari beberapa sub-atribut harus dipecah menjadi kolom terpisah.  
Contoh: `alamat â†’ jalan, kota, provinsi`

---

### **3. Atribut Multivalue**
Atribut bernilai banyak â†’ dibentuk sebagai **tabel baru**.

---

### **4. Entitas Lemah**
- Dibuat sebagai tabel baru.
- PK berupa gabungan PK entitas induk + atribut pembeda.

---

### **5. Relasi 1 â€” 1**
Foreign Key ditempatkan pada salah satu tabel, biasanya entitas yang lebih lemah atau yang lebih sering muncul.

---

### **6. Relasi 1 â€” N**
Foreign Key ditempatkan di sisi **N**.

---

### **7. Relasi 1 â€” N dengan Atribut Relasi**
Jika relasi memiliki atribut tambahan, harus dibuat **tabel relasional baru**.

---

### **8. Relasi N â€” M**
Selalu menghasilkan **tabel relasi**, yang berisi:
- FK entitas pertama  
- FK entitas kedua  
- Atribut relasi (jika ada)

---

### **9. Unary Relationship**
Relasi pada satu entitas:
- 1 â€” 1 â†’ FK pada tabel yang sama  
- N â€” M â†’ tabel relasi baru

---

### **10. Ternary Relationship**
Relasi tiga entitas menghasilkan:
- 3 tabel entitas  
- 1 tabel relasi

---

### **11. Generalisasi / Spesialisasi (ISA)**
Pendekatan konversi:
- Tabel superclass + tabel subclass  
- Hanya tabel subclass (semua atribut digabung)

---

### **12. Agregasi**
Relasi kompleks â†’ menghasilkan tabel relasi tambahan.
