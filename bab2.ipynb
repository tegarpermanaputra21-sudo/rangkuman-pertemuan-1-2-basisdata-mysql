## ðŸ”‘ Konsep Dasar

### **Basis Data**
Basis data adalah kumpulan data terstruktur yang disimpan secara sistematis agar:
- Mudah diakses  
- Mudah diperbarui  
- Mudah dikelola  
- Terhindar dari redundansi data  

Contoh: data siswa, data transaksi, data penjualan, dsb.

---

### **DBMS (Database Management System)**
DBMS adalah perangkat lunak yang digunakan untuk mengelola basis data.

Fungsi DBMS:
- Membuat database dan tabel
- Mengelola data (insert, update, delete)
- Mengatur keamanan data
- Menangani multi-user  
- Menjamin konsistensi data  

Contoh DBMS:
- MySQL  
- MariaDB  
- PostgreSQL  
- Oracle  
- SQL Server  

---

### **MySQL**
MySQL adalah DBMS populer bersifat open-source yang banyak digunakan untuk aplikasi web.

Kelebihan MySQL:
- Gratis dan open-source  
- Performa cepat & stabil  
- Mendukung bahasa SQL  
- Dukungan komunitas besar  
- Mudah dipasang dan digunakan
